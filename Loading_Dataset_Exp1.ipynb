# **Loading the Datasets**

*   [winequality-red.csv](https://archive.ics.uci.edu/dataset/186/wine+quality)
*   [winequality-white.csv](https://archive.ics.uci.edu/dataset/186/wine+quality)



```python
df1 = pd.read_csv('/content/winequality-red.csv')
df2 = pd.read_csv('/content/winequality-white.csv')
```

# **Methods:**

# **Methods:**

2.   **Concatenation**: pd.concat(dataframe1, dataframe2, ..., dataframen)



```python
df = pd.concat([df1, df2])
```

```python
df.head()
```

# **Methods:**

3.   **Merge**:
     merge() performs join operations similar to relational databases like SQL. Users who are familiar with SQL but new to pandas can reference a comparison with SQL.



```python
result = pd.merge(df1, df2, on='fixed acidity')
result.head()
```

# **Methods:**

4.   **Join**: DataFrame.join() combines the columns of multiple, potentially differently-indexed DataFrame into a single result DataFrame.

```python
result = df1.join(df2, lsuffix='_left', rsuffix='_right')
result.head()
```
